````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import seaborn as sns


files = glob.glob("states*.csv")

df_list = []
for filename in files:
  data = pd.read_csv(filename)
  df_list.append(data)

us_census = pd.concat(df_list)



````

## Regex to the Rescue

5. Use regex to turn the `Income` column into a format that is ready for conversion into a numerical type.

````python
#remove dollar sign
us_census['Income'] = us_census['Income'].replace('[$]', '', regex=True)
#remove commas and trimmin whitespace
us_census['Income'] = us_census['Income'].str.replace(',', '').str.strip()
#conerting to numeric
us_census['Income'] = pd.to_numeric(us_census['Income'])
#print(us_census.dtypes)
````

6. Look at the `GenderPop` column. We are going to want to separate this into two columns, the `Men` column, and the `Women` column.

   Split the column into those two new columns using `str.split` and separating out those results.

````python
# Split the string
string_split = us_census['GenderPop'].str.split('_')
 
#Men column
us_census['Men'] = string_split.str.get(0)
 
#Women column
us_census['Women'] = string_split.str.get(1)

print(us_census.head())
````

7. Convert both of the columns into numerical datatypes.

   There is still an `M` or an `F` character in each entry! We should remove those before we convert.

````python
#Replace any non-numeric values with NaN
us_census['Men'] = us_census['Men'].replace(r'[^\d.]', '', regex=True)
us_census['Women'] = us_census['Women'].replace(r'[^\d.]', '', regex=True)

#Removing M and F and converted to string
us_census['Men'] = us_census['Men'].astype(str).str.replace('M', '', regex=True)
us_census['Women'] = us_census['Women'].astype(str).str.replace('F', '', regex=True)

#Converting to numeric
us_census['Men'] = pd.to_numeric(us_census['Men'])
us_census['Women'] = pd.to_numeric(us_census['Women'])

print(us_census.dtypes)
print(us_census.head())
````

8. Now you should have the columns you need to make the graph and make sure your boss does not slam a ruler angrily on your desk because you've wasted your whole day cleaning your data with no results to show!

   Use matplotlib to make a scatterplot!
   
   ```py
   plt.scatter(the_women_column, the_income_column)
   ```
   
   Remember to call `plt.show()` to see the graph!

````python
#average income in a state vs proportion of women in that state.
plt.figure(figsize=(10, 10))
plt.scatter(us_census.Women, us_census.Income)
plt.xlabel('Proportion of Women')
plt.ylabel('Average Income')
plt.title('Scatterplot of Average Income vs Proportion of Women in States')
plt.show()
````

9. You want to double check your work. You know from experience that these monstrous csv files probably have `nan` values in them! Print out your column with the number of women per state to see.

   We can fill in those `nan`s by using pandas' `.fillna()` function.
   
   You have the `TotalPop` per state, and you have the `Men` per state. As an estimate for the `nan` values in the `Women` column, you could use the `TotalPop` of that state minus the `Men` for that state.
   
   Print out the `Women` column after filling the `nan` values to see if it worked!

````python
#print(us_census.Women)
values = {"Women": us_census.TotalPop - us_census.Men}
us_census['Women'] = us_census['Women'].fillna(value=values['Women'])
print(us_census.Women)
````

10. We forgot to check for duplicates! Use `.duplicated()` on your `census` DataFrame to see if we have duplicate rows in there.

````python
us_census.duplicated(subset = us_census.columns[1:]) #excluding the first columns
````

11. Drop those duplicates using the `.drop_duplicates()` function.

````python
us_census = us_census.drop_duplicates(subset = us_census.columns[1:])
us_census
````

12. Make the scatterplot again. Now, it should be perfect! Your job is secure, for now.

````python
plt.figure(figsize=(10, 10))
plt.scatter(us_census.Women, us_census.Income, alpha=0.5)
plt.xlabel('Proportion of Women')
plt.ylabel('Average Income')
plt.title('Scatterplot of Average Income vs Proportion of Women in States')
plt.show()
````

## Histogram of Races

13. Now your boss wants you to make a bunch of histograms out of the race data that you have. Look at the `.columns` again to see what the race categories are.

````python
us_census.columns
````

14. Try to make a histogram for each one!

    You will have to get the columns into the numerical format, and those percentage signs will have to go.
    
    Don't forget to fill the `nan` values with something that makes sense! You probably dropped the duplicate rows when making your last graph, but it couldn't hurt to check for duplicates again.

````python
#removing percentage signs
us_census['Hispanic'] = us_census['Hispanic'].replace('[%]', '', regex=True)
us_census['White'] = us_census['White'].replace('[%]', '', regex=True)
us_census['Black'] = us_census['Black'].replace('[%]', '', regex=True)
us_census['Native'] = us_census['Native'].replace('[%]', '', regex=True)
us_census['Asian'] = us_census['Asian'].replace('[%]', '', regex=True)
us_census['Pacific'] = us_census['Pacific'].replace('[%]', '', regex=True)
print(us_census.head())
#change to numeric
us_census['Hispanic'] = pd.to_numeric(us_census['Hispanic'])
us_census['White'] = pd.to_numeric(us_census['White'])
us_census['Black'] = pd.to_numeric(us_census['Black'])
us_census['Native'] = pd.to_numeric(us_census['Native'])
us_census['Asian'] = pd.to_numeric(us_census['Asian'])
us_census['Pacific'] = pd.to_numeric(us_census['Pacific'])
#print(us_census.dtypes)

#Fill nan values
#us_census.isna().sum() 
values = {"Pacific": 100 - (us_census.Hispanic + us_census.White + us_census.Black + us_census.Native + us_census.Asian)}
us_census['Pacific'] = us_census['Pacific'].fillna(value=values['Pacific'])
us_census.isna().sum() 
````

````python
#using a for loop
for race in ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']:
    for index in range(0,len(us_census)):     #iterating over each row in us_census
        string = str(us_census[race].iat[index])  #iat is used to access a single value for a row/column pair by integer position.
        replace = string.replace('%', '')
        if (replace == "nan"):
            replace = ""
        us_census[race].iat[index] = replace
    us_census[race] = pd.to_numeric(us_census[race])
    

````

````python
#histograms of race data

# List of race columns
race_columns = ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']  
for race in race_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(us_census[race])
    plt.title(f'Histogram of {race}')
    plt.xlabel('Percentage')
    plt.ylabel('Frequency')
    plt.show()
````

## Get Creative

15. Phew. You've definitely impressed your boss on your first day of work.

    But is there a way you really convey the power of pandas and Python over the drudgery of `csv` and Excel?
    
    Try to make some more interesting graphs to show your boss, and the world! You may need to clean the data even more to do it, or the cleaning you have already done may give you the ease of manipulation you've been searching for.

````python
#average income by race
races = ['Hispanic', 'White', 'Black', 'Asian', 'Native', 'Pacific']
weighted_income = {}
for race in races:
    weighted_income[race] = (us_census[race] * us_census['Income']).sum() / us_census[race].sum()

plt.bar(weighted_income.keys(), weighted_income.values()) #race names for the x axis and caluclted wighted income for the y
plt.title('Weighted Average Income by Race')
plt.xlabel('Race')
plt.ylabel('Average Income')
plt.show()
````
